```python
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile
from io import BytesIO
from xml.etree import ElementTree
```

```python
url='https://registers.esma.europa.eu/solr/esma_registers_firds_files/select?q=*&fq=publication_date:%5B2021-01-17T00:00:00Z+TO+2021-01-19T23:59:59Z%5D&wt=xml&indent=true&start=0&rows=100'
fil=requests.get(url)
soup = BeautifulSoup(fil.content, 'lxml')
print(soup.prettify())
```

```python
dtdiv=soup.find_all('doc',soup)
for di in dtdiv:
  if di.find('str', {'name': 'file_type'}).text == "DLTINS":
    url=di.find('str', {'name': 'download_link'}).text
    req=requests.get(url)
    zpf= ZipFile(BytesIO(req.content))
    zpf.extractall()
```

```python
for di in dtdiv:
  fn=di.find('str', {'name': 'file_name'}).text
  fn=fn.replace('.zip','.xml')
  tr=ElementTree.parse(fn)
  ro=tr.getroot()
  for instrm in ro.iter():
      print(instrm.tag)
  break
```
